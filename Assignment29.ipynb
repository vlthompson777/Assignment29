```python
#cluster racoon face
import numpy as np
from sklearn import cluster, datasets
from sklearn.cluster import KMeans
from scipy import misc
import matplotlib.pyplot as plt
from pylab import *
from scipy.spatial.distance import cdist
```

```python
racoon = misc.face(gray=True)
```

```python
plt.figure(figsize=(30, 10.18))
plt.subplot(131)
plt.imshow(racoon, cmap=plt.cm.gray)
```

```python
protos = array([[50,50],[100,200],[150,150],[350,700],[700,900]])
start = protos.copy()
```

```python
k_colors = KMeans(n_clusters=5).fit(racoon)
#Assign colors to pixels based on their cluster center
#Each row in k_colors.cluster_centers_ represents the RGB value of a cluster centroid
#k_colors.labels_ contains the cluster that a pixel is assigned to
#The following assigns every pixel the color of the centroid it is assigned to
fitimage=k_colors.cluster_centers_[k_colors.labels_]
```

```python
plt.imshow(fitimage)
```
